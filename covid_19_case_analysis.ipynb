```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("covid_19_dataset.csv")
```

```python
df.head()
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['AGE'], bins=30)
plt.title("Age Distribution of COVID Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='SEX', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
df['DEATH'] = df['DATE_DIED'].apply(
    lambda x: 0 if x == '9999-99-99' else 1
)

plt.figure(figsize=(6,4))
sns.countplot(x='DEATH', data=df)
plt.title("Death Distribution")
plt.xlabel("0 = Alive, 1 = Dead")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='DEATH', y='AGE', data=df)
plt.title("Age vs Death")
plt.show()
```

```python
pd.crosstab(df['DIABETES'], df['DEATH']).plot(kind='bar')

plt.title("Diabetes vs Death")
plt.xlabel("Diabetes")
plt.ylabel("Count")
plt.show()
```

```python
pd.crosstab(df['OBESITY'], df['DEATH']).plot(kind='bar')

plt.title("Obesity vs Death")
plt.xlabel("Obesity")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Age:", df['AGE'].mean())
print("Maximum Age:", df['AGE'].max())
print("Minimum Age:", df['AGE'].min())
```
